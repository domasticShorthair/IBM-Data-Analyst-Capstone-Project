## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(x=df['ConvertedComp'])
plt.title('Box Plot of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
q1 = df['ConvertedComp'].quantile(0.25)
q3 = df['ConvertedComp'].quantile(0.75)

inter_quartile_range = q3 - q1

print("q1:", q1)
print("q3:", q3)
print("Inter Quartile Range:", inter_quartile_range)
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = q1 - 1.5 * inter_quartile_range
upper_bound = q3 + 1.5 * inter_quartile_range
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
num_outliers = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)].shape[0]

print(f"Number of outliers in the 'ConvertedComp' column: {num_outliers}")
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_without_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
```
